# Aplicaciones de inferencia estadistica

## Datos de la universidad de graduados
Para realizar une studio estadistico a un set de datos debe realizarse a estos un pre-procesmaiento en donde los datos son preparados para efectuarles un analisis.

La idea es que si se sabe que materias le falta al estudiante se sabe cuando tiempo le puede faltar al estudiante para graduarse.

Para esto es un sistema que haga un aprendizaje reforzado para realizar el analisis como se desea.

Lo que se desea es efectuar un sistema de recomendación con la información de lo que le falta a los estudiantes para graduarse.


Algunos estudiantes han demorado mas de lo esperado pra graduarse, de hecho con una prueba estadistica sencilla para darse una idea del comportmaiento de los datos en un rango de semestres se puede establecer valores de media, desviación delimitado a carreras.

### Dandose una idea sobre los datos
Se delimita los datos a un rango de semestres y de cierta carrera para analizar el estudio en este caso el tiempo de semestres para graduarse obtenidno su media, desviación entre otros que serán utilizados para su posterior analisis

### Problemas
Este sistema de recomendación de estimar que tanto puede demorar un estudiante en graduarse cursando las materias en el orden que se le sugiera ha muchos factores que son relativos y variables ya que esta a razón de comportmaientos que tienen muchos factores que no hay forma de tenerlos presente como las condiciones medicas que le puedan surgir al estudiante, los problemas personales uqe tenga con algun docente entre otros.

### Predicción
A razón de los semestres que ha cursado y las materias que le esta faltando al estudiante se hace una predicción de cuantos semestres mas le va tomar graduarse

### Proceso
Se toma una muestra de estudiantes y a estos se le realiza la predicción a cada uno 5 veces almacenando todos sus resultados. de estos resultados se toman los mejores y son utilizados para comprobar que tan bueno es el sistema para hacer la predicción donde unos son de "entrenamiento" y otros sond e prueba

### Medias similares, datos similares
Se hace necesario determinar que tan similares son los datos entre si para saber si las muestras son iguales o diferentes.

Cada que se corre el algoritmo de predicción arroja datos brevemente distintos, la idea es verificar si son muy distintos o no. 

Al saber que si son iguales entre si permite saber si los datos del conjunto de datos se puede tomar el valor analizado ya que si son muy similares los datos con los demas del rupo de muestra puede usarse ese solo dato para comprar contra la población.

Mediante esto se puede establecer que tan consistentes son los datos, la idea e una prueba estadistica es que varie lo menos posible y entrege las "mismas" conclusioens con diferentes muestras.


### Predicción del tiempo de similitud
Calcular donde se encuentra ubicado semestralmente un estrudiante es clave para determinar bien el tiempo que le falta (recordemos que la ubicación semestra es a razón de que materias son las que tiene aprobadas independiente del tiempo que lleve en la carrera)

Las materias que tienen mas pre-requisitos son las que mayor peso tienen para el calculo de graduarse. El sistema de recomendación desarrollado por carlos andres lopez es capaz de explicar el por que va demorar el tiempo estimado

### Prueba beta o de potencia
La prueba beta que es la prueba de potencia permite conocer esos parametros de cuando el sistema es correcto. Es decir determina a su ves que tanto puede fallar el sistema respecto a los resultados que ha dado hasta el momento, la idea es determina esos valores criticos donde la prueba tiende a fallar mas
