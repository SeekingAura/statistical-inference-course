## Hiptesis alternativa
Generlamente la hipotesis alternativa $H_{a}$ es la que se quiere apoyar o demostrar y la hipotesis nula $H_{0}$ es la contradicción de la hipotesis alternativa, es decir que se trata de una prueba de reducción al absurdo

### Puede presentarse lo siguiente
* Si se rechaza $H_{0}$ se puede concluir que $H_{a}$ es verdadera
* Acepta (no rechazo) de $H_{0}$ como verdadera

La logica difusa hace funcionamiento en estas circunstnatancias. Las hipostesis en si se trata de plantear algo con su aprte positiva y su parte negativa (o sea dos planteamientos donde se indic alago y otro donde se niega el otro

## Ejemplo
Se desea demostrar que el promedio salarial en el empleo informal es diferente del salario mínimo legal vigente, entonces:

$H_{a} : \mu \neq SMLV$
$H_{0} : \mu = SMLV$

Si se rechaza la hipótesis nula se demeustra que se concluye que el salario no es igual al SMLV

## Ejemplo 2
El promedio de deserción de una universidad para un programa es de 65% de estudiantes, se desea demostrar que con un ajuste en la politica de ingreso reducirá p, la proporción de estudiantes que desertan, entonces:
* $H_{a} : p<0.65$
* $H_{0} : p>0.65$

# Diferencias entre pruebas de hipotesis
* En el primer ejemplo se le llama **prueba de hipostesis de dos colas**
* El el segundo ejemplo se busca encontrar una diferencia direccional respecto al valor de p, es decir que si $H_{a}$ es verdadera el valor de p es menor al 65%. Este tipo de prueba se llama **prueba de hipotesis de uan cola**

# Datos de la prueba de hipotesis
Rechazar o aceptar la hipótesis nula se fundamenta en la información de la muestra extraída de la población de interés y aparece como:
* Estadística de prueba: un solo valor

## Regiones
Para aceptar una hipótesis se revisan las regiones de aceptación o rechazo. Los valores que apoyyan la hipótesis alternativa $H_{a}$ se denomina región de rechazo, los valores que apoyan la hipotesis nula se le denomina región de aceptación

### Las regiones
En una región de rechazo de dos colas se buscan valores tal que la medida es mucho mayor o mucho menor respecto a esta
```
____Rechazo____|___Aceptación___|___Rechazo___
```

Cuando una región de rechazo se encuentra en cola izqueirda de distribución se llama **prueba de cola izquierda**

```
____065____|___rechazo
```

# Valores críticos

# Método de valor
El **valor p** prueba de significación es el valor mas pequeño de $\alpha$ para el cual $H_{0}$ se puede rechazar

* Un **valor p pequeño** es una fuerte evidencia que $H_{0}$ es falsa y se peude rechazar
* Un **valor de p grande** indica que la media no esta alejada de la media hipotética y no apoya el rechazo de $H_{0}$. En este caso indica que deberia quizas replantearse la hipotesis

## Escala del valor p
* Si el valor es menor a 0.01 $H_{0}$ se rechaza y los resultados son altamente significativos
* Si el valor p está entre 0.01 y 0.05 $H_{0}$ se rechaza y los reusltados son estadisticamente significativos (la mayoria de estos resultan asi)
* Si el valor p está entre 0.05 y 0.10 H_{0} por lo general no se rechaza y resultados tendientes hacia significancia estadistisca
* si el valor p es mayor a 0.10 $H_{0}$ no se recahza y los resultados no son estadisticmaente significativos (en este caso el algoritmo debe ser ajustado si es en machine learning)

## Calculo de valor p
Se $z_p$


## Ejemplo
En el ejemplo de la trilladora con producción promedio de 880 toneladas y para una meustra de 50 datos con promedio de 871 toneladas y desviación estándar de 21 toneladas calcular el valor p.

Como el valor observado el estadístico de prueba z=-3.03 la región de rechazo:

p=P(z>3.03)+P(<-3.03)=(1-0.9988)+0.0012=0.0024

$H_{0}$ puede ser rechazada con nivel de significancia de 1% y 5% siendo estadisticamente sigfnificativos
