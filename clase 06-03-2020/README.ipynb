# Probabilidad condicional
Es la probabilidad de ocurrencia de un evento dado a que otro ocurra, se expresa como

$P(A|B)=\frac{P(A\cap B)}{P(B)}$

## ejemplo
La fuerza policiaca consta de 1200 agentes, 960 hombres y 240 mujeres. De estos, en los últimos dos años, fueron promovidos 340. En la tabla se muestran estas promociones entre os hombres y mujeres

|            | Hombre | Mujer | Total |
|------------|:------:|:-----:|:-----:|
|Promovido   |288     | 36    | 324   |
|No promovido| 627    | 204   | 876   |
| Total      | 960    | 240   | 1200  |

La proporción marginal es: Es la proporción que hay entre unos y otros
La tabla marginal es:
|            | Hombre | Mujer | Total |
|------------|:------:|:-----:|:-----:|
|Promovido   | 0.24 ->   | 0.03 ->    | 0.27  |
|No promovido| 0.56    | 0.17 -> | 0.73   |
| Total      | 0.80    | 0.20   | 1.0  |

Supongamos que **M** es que el agente sea promovido, **A** sea hombre y **W** sea mujer, para calcular la probabildiad de que un hombre sea promovido es:

$P(M\cap A)=\frac{288}{1200}=0,24$
$P(M\cap W)= 0.03$ 
$P(M'\cap A)=0.56$
$P(M'\cap W)=0.20$ 

Probabilidad de que un hombre sea promovido

* $P(A|M)=\frac{A\cap M}{P(M)}=\frac{0.24}{0.80}=0.30$




Si los datos no tienen ningún tipo de relación se da que P(A|B)=P(A)

Si los eventos estan realcionados aplica las multiplicaciones y sumas aplicadas
En este caos los eventos están mezclados y tienen dependencia

Cuando el coeficiente de perason es 0 no hay regresión lineal
